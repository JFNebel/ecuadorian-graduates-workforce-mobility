```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/clean_supercias_links_mainDF.csv')
```

```python
df.info()
```

```python
df['ACTIVIDAD ECONÓMICA'].nunique()
```

```python
industries = [
    "Farmaceutica",
    "Manufactura",
    "Automotriz",
    "Agricultura/Agropecuaria",
    "Telecomunicaciones/Informática",
    "Logística/Transporte",
    "Minería",
    "Minoristas",
    "Mayoristas",
    "Construcción",
    "Otro"
]
```

```python
print(*enumerate(industries), sep='\n')
```

```python
df_industries = pd.DataFrame(df['ACTIVIDAD ECONÓMICA'].unique())
```

```python

for index, row in df_industries.iterrows():
    print(*enumerate(industries), sep='\n')
    print(row[0])
    opt = int(input("Select the option: "))
    df_industries.loc[index, "INDUSTRIA"] = industries[opt]
```

```python
df_industries
```

```python
df_industries['INDUSTRIA'].nunique()
```

```python
df_industries['INDUSTRIA'].value_counts()
```

```python
df_industries.columns = ['ACTIVIDAD ECONÓMICA', 'INDUSTRIA']
```

```python
final_df = pd.merge(df, df_industries, how="left", on=["ACTIVIDAD ECONÓMICA"])
```

```python
final_df.groupby('INDUSTRIA')['NOMBRE'].count()
```

```python
final_df.to_csv('./Data/companies_with_industries.csv')
```
