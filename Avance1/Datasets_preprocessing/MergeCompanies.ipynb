```python
import pandas as pd
```

```python
top_100_links = pd.read_csv('Data/top_100.csv')
supercias_df = pd.read_csv('Data/rankingSuperCias.csv', sep=';', encoding='latin') 
supercias_df.head()
```

```python
supercias_df.dropna(inplace=True)
supercias_df.shape
```

```python
# Use subset of SuperCias dataset to merge with top_100 dataframe
supercias_df = supercias_df.head(345)
```

```python
# There is unwanted whitespace in the "NOMBRE" column let's clean it
def clean_space(row):
  return row.strip()

supercias_df.NOMBRE = supercias_df.NOMBRE.apply(clean_space)
top_100_links.NOMBRE = top_100_links.NOMBRE.apply(clean_space)
```
