# Cleaning the \n in the dataset

```python
def clean_denominacion(row):
    row['DENOMINACION'] = row['DENOMINACION'].strip().replace('\n', ' ')
    return row
```

```python
df_universities = df_universities.apply(clean_denominacion, axis=1)
```

```python
df_uni_location = df_universities[['DENOMINACION','PROVINCIA', 'CIUDAD']]
```
