```python
import pandas as pd
df = pd.read_csv('Data/what_they_are_skilled_at.csv')
```

```python
df = df.fillna(0)
```

```python
df = df.iloc[:,2:]
```

```python
df = df.convert_dtypes()  
```

```python
df['Customer Service']= df['Customer Service'].str.replace(',','')
df['Customer Service'] = pd.to_numeric(df['Customer Service'])
```

```python
df['Leadership']= df['Leadership'].str.replace(',','')
df['Leadership'] = pd.to_numeric(df['Leadership'])
```

```python
df['Teamwork']= df['Teamwork'].str.replace(',','')
df['Teamwork'] = pd.to_numeric(df['Teamwork'])
```

```python
df['English']= df['English'].str.replace(',','')
df['English'] = pd.to_numeric(df['English'])
```

```python
df_corr = df.corr()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.matshow(df_corr)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(df_corr, annot=True)
plt.show()
```
