```python
#Si se calcula la relación del número de veces que aparece cada palabra en 
# comparación con el número de palabras del documento, obtenemos lo siguiente.
# Fórmulas: max, sum, 1 + log(freq)

def applytf(s):
    return s/s.sum()

dftf=df_doc.apply(applytf,axis=1)
dftf
```

```python
#idf 
def applyidf(s):    
    d = len(s)
    n = (s>0).values.sum()
    n = (n if (n>0) else 1)
    return log(d/(n))
    #return log(d/(n+1))

dfidf = df_doc.apply(applyidf,axis=0)
dfidf = pd.DataFrame(dfidf)
dfidf = dfidf.transpose()
dfidf
```

```python
x = df_final
wcss = []
for i in range(1,25):
    kmeans = KMeans(n_clusters= i , init = 'k-means++' , random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.style.use(['dark_background'])
plt.plot(range(1,25) , wcss)
plt.title('Optimal No.of clusters VS WCSS')
plt.xlabel
```

```python
visual = KElbowVisualizer(kmeans, k=(2, 73))
visual.fit(df_final)
visual.elbow_value_
```
